```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
bank=pd.read_csv("bank-full.csv",sep=";")
bank.head()
```

```python
# shape of dataset
bank.shape
```

```python
# check na value
bank.isna().sum()
```

```python
bank.dtypes
```

###### new dataset

```python
# as per dtypes, the categorical data must to convert into numeric data

bank['job_c']=pd.factorize(bank.job)[0]
bank['marital_c']=pd.factorize(bank.marital)[0]
bank['education_c']=pd.factorize(bank.education)[0]
bank['default _c']=pd.factorize(bank.default )[0]
bank['housing _c']=pd.factorize(bank.housing )[0]
bank['loan_c']=pd.factorize(bank.loan)[0]
bank['contact_c']=pd.factorize(bank.contact)[0]
bank['month_c']=pd.factorize(bank.month)[0]
bank['poutcome_c']=pd.factorize(bank.poutcome)[0]
bank['y_c']=pd.factorize(bank.y)[0]

bank_new1 = bank
bank_new1
```

```python
# also need to drop series column which convert to numeric
bank.drop(['job','marital','education','default','housing','loan','contact','month','poutcome','y'], inplace=True,axis = 1)
bank_new = bank
```

```python
# new dataset in numeric form
bank_new
```

```python
# Dividing our data into input and output variables 
# input
X = bank_new.iloc[:,0:16]
X
```

```python
# output
Y = bank_new.iloc[:,16]
Y
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```python
#Predict for X dataset
Y_pred = classifier.predict(X)
Y_pred
```

```python
y_pred_df= pd.DataFrame({'actual': Y,'predicted_prob':Y_pred})
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,Y_pred)
print (confusion_matrix)
```

###### Classification report

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(Y,Y_pred))
```

```python
# ROC Curve (receiver operating characteristic curve)

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, classifier.predict_proba (X)[:,1])

auc = roc_auc_score(Y, Y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
